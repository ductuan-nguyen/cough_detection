```python
import librosa
import os
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
DATA_NO_COUGH = 'audio_1s'
DATA_COUGH = 'data_1s'
```

```python
file_cough_list = os.listdir(DATA_COUGH)
file_nocough_list = os.listdir(DATA_NO_COUGH)
```

```python
file_cough_list = os.listdir(DATA_COUGH)
i = 1
plt.figure(figsize=(21, 25))
for filename in file_cough_list[30:45]:
    y, sr = librosa.load(os.path.join(DATA_COUGH, filename))
    ax = plt.subplot(5, 3, i)
    ax.set_title(filename, color='green')
    librosa.display.waveplot(y, sr=sr)
    i+=1
plt.show()
```

```python
file_nocough_list = os.listdir(DATA_NO_COUGH)
i = 1
plt.figure(figsize=(21, 25))
for filename in file_nocough_list[30:45]:
    y, sr = librosa.load(os.path.join(DATA_NO_COUGH, filename))
    ax = plt.subplot(5, 3, i)
    ax.set_title(filename, color='green')
    librosa.display.waveplot(y, sr=sr)
    i+=1
plt.show()
```

```python
i = 1
plt.figure(figsize=(28, 25))
for filename in file_cough_list[0:20]:
    y, sr = librosa.load(os.path.join(DATA_COUGH, filename))
    ax = plt.subplot(5, 4, i)
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=256, fmax=16000)
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, fmax=16000, ax=ax)
    # fig.colorbar(img, ax=ax, format='%+2.0f dB')
    ax.set(title='Mel-frequency spectrogram')
    i+=1
plt.show()
```

```python
i = 1
plt.figure(figsize=(28, 25))
for filename in file_nocough_list[0:20]:
    y, sr = librosa.load(os.path.join(DATA_NO_COUGH, filename))
    ax = plt.subplot(5, 4, i)
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=256, fmax=16000)
    S_dB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, fmax=16000, ax=ax)
    # fig.colorbar(img, ax=ax, format='%+2.0f dB')
    ax.set(title='Mel-frequency spectrogram')
    i+=1
plt.show()
```
